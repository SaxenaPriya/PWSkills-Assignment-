# **Theory Questions**
1. What is Python, and why is it popular?
- Python is a high-level, interpreted programming language known for its simple syntax and readability. It is popular because it is easy to learn, versatile, supports multiple programming paradigms, and has a large collection of libraries for web development, data science, AI, and automation.

2. What is an interpreter in Python?
- For Python, an interpreter is a program that executes code line by line. It converts source code into machine code at runtime, which makes debugging easier.

3. What are pre-defined keywords in Python?
- Keywords are reserved words that have special meanings in Python and cannot be used for other purposes.
Example: if, else, for, while, True, False, class.

4. Can keywords be used as variable names?
- No, Python keywords cannot be used as variable names because they are reserved for specific operations only.

5. What is mutability in Python?
- Mutability refers to whether an objectâ€™s value can be changed after it is created.

6. Why are lists mutable, but tuples are immutable?
- Lists are designed to allow changes like adding or removing elements, while tuples are meant to be fixed, making them faster and safer for constant data.

7. What is the difference between == and is operators in Python?
- "==" compares values for equality, while is checks whether both variables refer to the same object in memory location.

8. What are logical operators in Python?
- Logical operators are used to combine conditional statements.
Examples: and, or, not.

9. What is type casting in Python?
- Type casting is the process of converting one data type into another.
Example: converting int to float, float to int, input to int or float,etc.

10. What is the difference between implicit and explicit type casting?

- Implicit casting is where we need to define the data type but done automatically in Python.

- Explicit casting is converting one datatype into another by manually using functions like int(), float(), str().

11. What is the purpose of conditional statements in Python?
- Conditional statements control the flow of a program by executing code only when given certain conditions are met or matches.

12. How does the elif statement work?
- elif checks another condition if the previous if condition is false. It allows multiple condition checks in sequence.

13. What is the difference between for and while loops?
- A for loop runs for a fixed number of iterations, while a while loop runs as long as a condition remains true.

14. Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is suitable when the number of iterations is unknown, such as reading user input until the correct password is entered.

```python
"""
6. Write a function to demonstrate the behavior of mutable and immutable arguments.
"""
def demo_mutable(a_list):
  print("Inside before:", a_list)
  a_list.append(100)
  print("Inside after:", a_list)

def demo_immutable(a_tuple):
  print("Inside before:", a_tuple)
  try:
    a_tuple += (99,)
    print("Inside after:", a_tuple) # new object added
  except Exception as e:
    print("Error:", e)

a_tuple = (1, 2, 3, 4, 5)
a_list = [1, 2, 3, 4, 5]
demo_immutable(a_tuple)
print()
demo_mutable(a_list)
```

```python
"""
14. Write a program to reverse a string using a while loop.
"""
s = input("Enter a string: ")
rev = ""
i = len(s) - 1

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed string:", rev)
```

```python
"""
15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
"""
num = int(input("Enter a number: "))
fact = 1

while num > 0:
    fact *= num
    num -= 1

print("Factorial:", fact)
```
